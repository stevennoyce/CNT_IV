```python
import os
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('../')

from utilities import DataLoggerUtility as dlu
from utilities import DataPlotterUtility as dpu
import B2912A_Main_User_Interface_Script as mainUI

from matplotlib import pyplot as plt
import numpy as np

# %config InlineBackend.figure_format = 'svg'


def makePlots(waferID, chipID, deviceID, startExperimentNumber=0, endExperimentNumber=float('inf'), 
              specificPlot='', figureSize=None, saveFolder=None, saveNamePrefix = '', save=False, 
              startRelativeIndex=0, endRelativeIndex=float('inf')):
    parameters = mainUI.default_parameters
    
    dpu.publication_mode = True
    
    deviceDirectory = 'data/' + waferID + '/' + chipID + '/' + deviceID
    indexes = dlu.loadIndexesOfExperiementRange(deviceDirectory, startExperimentNumber, endExperimentNumber)
    
    parameters['runType'] = 'DeviceHistory'
    parameters['waferID'] = waferID
    parameters['chipID'] = chipID
    parameters['deviceID'] = deviceID
    parameters['DeviceHistory']['excludeDataBeforeJSONExperimentNumber'] = startExperimentNumber
    parameters['DeviceHistory']['excludeDataAfterJSONExperimentNumber'] = endExperimentNumber
    parameters['DeviceHistory']['excludeDataBeforeJSONIndex'] = min(indexes) + startRelativeIndex
    parameters['DeviceHistory']['excludeDataAfterJSONIndex'] = min(indexes) + endRelativeIndex
    parameters['DeviceHistory']['postFiguresGenerated'] = False
    parameters['DeviceHistory']['saveFiguresGenerated'] = save
    parameters['DeviceHistory']['gateSweepDirection'] = 'reverse'
    parameters['DeviceHistory']['plotInRealTime'] = True
    parameters['DeviceHistory']['specificPlotToCreate'] = specificPlot
    parameters['DeviceHistory']['figureSizeOverride'] = figureSize
    if saveFolder is not None:
        parameters['plotsFolder'] = saveFolder + '/'
    
    parameters['plotsFolder'] += saveNamePrefix
    
    mainUI.runAction(parameters, None, None)


saveFolder = '../../../Publications/2018-04-30 Bias Stress ACS/Plots'
```

```python
print('# Figure 3a')

staticBiass = dlu.loadJSON('data/C127/E/15-16/', 'StaticBias.json')
staticBiass = dlu.filterHistoryGreaterThan(staticBiass, 'experimentNumber', 3)
staticBiass = dlu.filterHistoryLessThan(staticBiass, 'experimentNumber', 10)

gateSweeps = dlu.loadJSON('data/C127/E/15-16/', 'GateSweep.json')
gateSweeps = dlu.filterHistoryGreaterThan(gateSweeps, 'experimentNumber', 3)
gateSweeps = dlu.filterHistoryLessThan(gateSweeps, 'experimentNumber', 10)

initialTime = staticBiass[0]['Results']['timestamps'][0]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(2.22,1.6), sharex=True, gridspec_kw={'width_ratios':[1], 'height_ratios':[3,1]})
# fig = plt.figure(figsize=(2,1.6))
onCurrents = []
times = []
timeStampss = []
VDSs = []

for gateSweep, staticBias in zip(gateSweeps, staticBiass):
    timestamps = np.array(staticBias['Results']['timestamps'])/3600/24
    times.append(timestamps[0])
    timeStampss.append(timestamps - times[0])
    VDSs.append(staticBias['StaticBias']['drainVoltageSetPoint'])
#     Id = np.array(staticBias['Results']['id_data'])
    onCurrents.append(gateSweep['Results']['onCurrent'])
    ax2.set_prop_cycle(None)
    ax2.plot([])
    ax2.plot((timestamps - times[0]), np.ones(len(timestamps))*staticBias['StaticBias']['drainVoltageSetPoint'])


colors = dpu.colorsFromMap('hot', 0.7, 0, len(times))['colors']

times = np.array(times)
times = times - times[0]
onCurrents = np.array(onCurrents)
ax1.scatter(times, abs(onCurrents)*1e6, s=2, c=colors)

# [tick.set_verticalalignment('baseline') for tick in ax1.yaxis.get_majorticklabels()]
ax1.yaxis.get_majorticklabels()[0].set_verticalalignment('baseline')
ax1.legend(title='$V_{{GS}}^{{Hold}}$ = {} V'.format(staticBiass[0]['StaticBias']['gateVoltageSetPoint']), labelspacing=0)
ax1.set_ylim(ymin=0)
ax2.set_xlabel('Time [Days]')
ax2.set_ylabel('$V_{{DS}}^{{Hold}}$ [V]')
ax1.set_ylabel('$I_D$ [$\mu$A]')
plt.tight_layout()
fig.subplots_adjust(hspace=0.03)
fig.align_labels()
plt.savefig(saveFolder + '/Figure 3a - Custom OnCurrent.png', transparent=True, dpi=300)
plt.show()
```

```python
from scipy.optimize import curve_fit

def trapFill1(t, v, A, tau):
    return v + A*np.exp(-t/tau)

popts = []

for parameters in parametersSets:
    timestamps = np.array(parameters['Results']['timestamps'])
    timestamps = timestamps - timestamps[0]
    timestamps = timestamps/3600
    Id = np.array(parameters['Results']['id_data'])
    
    popt, pcov = curve_fit(trapFill1, timestamps, abs(Id))
    
    popts.append(popt)
    
    timestamps2 = np.linspace(0,2,100)
    plt.plot(timestamps2, trapFill1(timestamps2, *popt))
    plt.plot(timestamps, abs(Id))

plt.show()

taus = [popt[2] for popt in popts]
plt.semilogy(taus)
plt.ylabel('Tau [hours]')
plt.show()

print(np.median(taus))
```

```python
from scipy.optimize import curve_fit

def trapFill2(t, v, A):
    tau = 0.244146722649
    tau = 0.5
    return v + A*np.exp(-t/tau)

for parameters in parametersSets:
    timestamps = np.array(parameters['Results']['timestamps'])
    timestamps = timestamps - timestamps[0]
    timestamps = timestamps/3600
    Id = np.array(parameters['Results']['id_data'])
    
    popt, pcov = curve_fit(trapFill2, timestamps, abs(Id))
    
    timestamps2 = np.linspace(0,2,100)
    plt.plot(timestamps2, trapFill2(timestamps2, *popt))
    plt.plot(timestamps, abs(Id))

plt.show()
```

```python
IdsInitial = []
IdsFinal = []
Vgss = []

for parameters in parametersSets:
    Id = np.array(parameters['Results']['id_data'])
    Vgs = np.array(parameters['Results']['vgs_data'])
    IdsInitial.append(np.mean(Id[0:2]))
    IdsFinal.append(np.mean(Id[-60:-1]))
    Vgss.append(Vgs[0])

fig = plt.figure(figsize=(1.7,1.89))
colors = dpu.colorsFromMap('plasma', 0, 0.75, 2)['colors']
plt.semilogy(Vgss, np.abs(IdsInitial), label='Initial\n   DR = {:.0f}\n'.format(IdsInitial[-1]/IdsInitial[0]), c=colors[0])
plt.semilogy(Vgss, np.abs(IdsFinal), label='Terminal\n   DR = {:.0f}'.format(IdsFinal[-1]/IdsFinal[0]), c=colors[1])
plt.legend()
plt.xlabel('$V_{GS}^{{Hold}}$')
plt.ylabel('$I_D$ [A]')
plt.tight_layout()
plt.savefig(saveFolder + '/Figure 5c - Initial and Terminal Responses.png', transparent=True, dpi=300)
plt.show()
```
